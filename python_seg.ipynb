```python
import pandas as pd

tabela = pd.read_csv("/content/drive/MyDrive/projeto analise data/Airline_customer_satisfaction.csv.zip", encoding="latin1")

tabela = tabela.drop("Departure Delay in Minutes", axis=1)
tabela = tabela.drop("Arrival Delay in Minutes", axis=1)


display(tabela)

```

```python
tabela = tabela.dropna()
display(tabela.info())

display(tabela.describe(). round(1))
```

```python
qtd_sati = tabela["satisfaction"]. value_counts()
qtd_comfort = tabela["Seat comfort"]. value_counts()
qtd_food = tabela["Food and drink"]. value_counts()
qtd_distance = tabela["Flight Distance"]. value_counts()
qtd_age = tabela["Age"]. value_counts()


display(qtd_sati)
display(qtd_comfort)
display(qtd_food)
display(qtd_distance)
display(qtd_age)
```

```python
import plotly.express as px
for coluna in tabela:
  grafico = px.histogram(tabela, x=coluna, color="satisfaction")
grafico.show()
```
